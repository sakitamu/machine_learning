## 5_4.独立した2群の差のt検定(分散が等しいと仮定しない場合のウェルチのt検定)

```python
import pandas as pd

#  CSVファイルのデータをデータフレームに読み込む
df = pd.read_csv('research.csv')
# 基本統計量を出力
df.describe()
```

```python
from scipy import stats

# ウェルチのt検定を実施
# nan_policy='omitでNaNを無視して計算を行う
statistic, pvalue = stats.ttest_ind(
    df['A'],
    df['B'],
    nan_policy='omit',
    equal_var=False
)

# 結果を出力
print('検定統計量： ', statistic)
print('p値：　', pvalue)
```
