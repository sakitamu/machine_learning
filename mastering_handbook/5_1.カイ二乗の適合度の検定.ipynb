## 5_1.カイ二乗の適合度の検定

```python
from scipy import stats

# サイコロ投げ（２０回における１の目から６の目の観測度数）
observed = [3, 1, 0, 2, 4, 2]

# １の目から６の目の期待度数
expected = [2, 2, 2, 2, 2, 2]

# 検定統計量とp値を求める
chisq, p = stats.chisquare(observed, expected)
print('検定統計量：　', chisq)
print('p値：　', p)
```
