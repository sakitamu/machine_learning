## 4_2.散布図と相関関係


```python
import pandas as pd

# CSVファイルのデータをデータフレームに読み込む
df = pd.read_csv(
    # 同じフォルダ内のCSVファイルを読み込む
    'sales.csv'
)
```

```python
import matplotlib.pyplot as plt

# 売上数を縦軸，最高気温を横軸に設定して散布図を描画
df.plot(
    # 横軸は最高気温(列インデックス:0)
    x=df.columns[0],
    # 縦軸は売上数（列インデックス:1)
    y=df.columns[1],
    # グラフの種類は散布図
    kind='scatter',
    # x軸ラベル
    xlabel='Highest temprature',
    # y軸ラベル
    ylabel='Sales'
)

# 散布図を出力
plt.show()
```

```python
# '最高気温'列のデータをNumpyに配列として取得
x = df['最高気温'].values
# '清涼飲料水売上数'列のデータをNumpyに配列として取得
y = df['清涼飲料売上数'].values

# '最高気温'の平均値を求める
mean_x = x.sum() / len(x)
# '清涼飲料水売上数'の平均値を求める
mean_y = y.sum() / len(y)

# xとyの共分散を求める
sxy = ((x - mean_x)*(y - mean_y)).sum() / len(x)
# 共分散を出力
sxy
```

```python
import numpy as np

# cov()関数で共分散を求める
# bias=Trueで標本分散を指定
sxy = np.cov(x, y, bias=True)
# 出力を指数表記ではなく小数にする
np.set_printoptions(suppress=True)
# 共分散を出力
sxy
```

```python
# xの標準偏差
sx = np.sqrt(((x - mean_x)**2).sum() / len(x))
# yの標準偏差
sy = np.sqrt(((y - mean_y)**2).sum() / len(y))
# xとyの共分散
sxy = ((x - mean_x)*(y - -mean_y)).sum() / len(x)

# x,yの相関係数を求める
sxy / (sx * sy)
```

```python
# NumPyのcorrcoef()で相関係数を求める
np.corrcoef(x, y)
```
