## 5_6.3群の平均に差があるかを分散分析で知る

```python
import pandas as pd
from scipy import stats

# CSVファイルのデータをデータフレームに読み込む
df = pd.read_csv('practice_test.csv')

# 対応のない３群の分散分析を実施
statistic, pvalue = stats.f_oneway(df['course_A'], df['course_B'], df['course_C'])

# 結果を出力
print('検定統計量：　', statistic)
print('p値：　', pvalue)
```
