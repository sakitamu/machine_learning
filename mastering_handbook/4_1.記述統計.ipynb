## 4_1.記述統計


```python
import pandas as pd

# CSVファイルのデータをデータフレームに読み込む
df = pd.read_csv(
    # 同じフォルダ内のCSVファイルを読み込む
    'access.csv',
    # 見出し列のインデックスを指定
    index_col=0
)
# データフレームの冒頭5件を出力
df.head()
```

```python
# 'accesses'列のデータをNumpy配列として取得
data = df['accesses'].values
# ３０日間のアクセス数の平均値を求める
mean = data.sum() / len(data)
# 平均値を出力
mean
```

```python
# Numpyをインポート
import numpy as np
# mean()関数で平均値を求める
np.mean(data)
```

```python
# 最小値を求める
np.min(data)
```

```python
# 最大値を求める
np.max(data)
```

```python
# 中央値を求める
np.median(data)
```

```python
# 30日間のアクセス数の平均値を求める
mean = data.sum() / len(data)
# 分散を求める
((data - mean)**2).sum() / len(data)
```

```python
# NumPyのvar()関数で分散を求める
np.var(data)
```

```python
# 30日間のアクセス数の標準偏差を求める
np.sqrt(
    ((data - mean)**2).sum() / len(data)
)
```

```python
# NumPyのstd()関数で標準偏差を求める
np.std(data)
```
